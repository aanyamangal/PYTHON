```python
import pandas as pd

# Load the weather dataset
df = pd.read_csv("weather.csv")  

# Display first few rows
df.head()
```

```python
df.columns
```

```python
# Fix Date conversion
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
# Task 3: Statistical Analysis

print("Basic Statistics:\n")
print(df[['Temperature', 'Humidity', 'WindSpeed', 'Pressure']].describe())

print("\nMonthly Average Temperature:\n")
print(df.groupby('Month')['Temperature'].mean())
```

```python
# Task 4: Visualization with Matplotlib
import matplotlib.pyplot as plt

# 1 Line Chart - Daily Temperature Trend
plt.figure(figsize=(10,4))
plt.plot(df['Date'], df['Temperature'])
plt.title("Daily Temperature Trend in Delhi")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.savefig("daily_temperature_trend.png")
plt.show()

# 2 Bar Chart - Monthly Average Temperature
monthly_temp = df.groupby('Month')['Temperature'].mean()
plt.figure(figsize=(8,4))
monthly_temp.plot(kind='bar')
plt.title("Monthly Average Temperature in Delhi")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.savefig("monthly_avg_temperature.png")
plt.show()

# 3 Scatter Plot - Humidity vs Temperature
plt.figure(figsize=(6,4))
plt.scatter(df['Temperature'], df['Humidity'])
plt.title("Humidity vs Temperature in Delhi")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.savefig("temp_vs_humidity.png")
plt.show()
```

```python
# Task 5: Export cleaned dataset
df.to_csv("cleaned_weather_data.csv", index=False)
print("Cleaned dataset saved as cleaned_weather_data.csv")
```
